```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Generate synthetic data with different feature scales
np.random.seed(0)

# Daily temperature (feature x1) with a range of [20, 40] degrees Celsius
temperature = 20 + 20 * np.random.rand(100, 1)

# Ice cream sales (feature x2) with a range of [0, 200] units
ice_cream_sales = 200 * np.random.rand(100, 1)

# Combine temperature and ice cream sales into a single feature matrix X
X = np.hstack((temperature, ice_cream_sales))

# Daily profit (target variable y)
daily_profit = 1000 + 30 * temperature + 5 * ice_cream_sales + np.random.randn(100, 1)

# Define the function for Max Scaling
def max_scale(X):
    return X / X.max(axis=0)

# Define the function for Mean Scaling
def mean_scale(X):
    return (X - X.mean(axis=0)) / X.std(axis=0)

# Define the function for Z-Score Scaling (Standardization)
def z_score_scale(X):
    return StandardScaler().fit_transform(X)

# Scale the features using the three methods
X_max_scaled = max_scale(X.copy())
X_mean_scaled = mean_scale(X.copy())
X_z_score_scaled = z_score_scale(X.copy())

# Function for cost calculation (Mean Squared Error)
def calculate_mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Rest of the code for linear regression, gradient descent, and evaluation

# Define a function for gradient descent
def gradient_descent(X, y, alpha, num_iterations):
    m, n = X.shape
    w = np.zeros((n, 1))
    b = 0

    for iteration in range(num_iterations):
        y_pred = np.dot(X, w) + b
        w_gradient = (1 / m) * np.dot(X.T, (y_pred - y))
        b_gradient = (1 / m) * np.sum(y_pred - y)
        w -= alpha * w_gradient.reshape(-1, 1)
        b -= alpha * b_gradient

    return w, b

# Define a function for making predictions
def predict(X, w, b):
    return np.dot(X, w) + b

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_max_scaled, daily_profit, test_size=0.2, random_state=42)

# Train the model using gradient descent
w_max_scaled, b_max_scaled = gradient_descent(X_train, y_train, alpha=0.01, num_iterations=1000)

# Make predictions on the test data
y_pred_max_scaled = predict(X_test, w_max_scaled, b_max_scaled)

#######################################################################################################
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_mean_scaled, daily_profit, test_size=0.2, random_state=42)

# Train the model using gradient descent
w_mean_scaled, b_mean_scaled = gradient_descent(X_train, y_train, alpha=0.01, num_iterations=1000)

# Make predictions on the test data
y_pred_mean_scaled = predict(X_test, w_mean_scaled, b_mean_scaled)

########################################################################################################

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_z_score_scaled, daily_profit, test_size=0.2, random_state=42)

# Train the model using gradient descent
w_z_score_scaled, b_z_score_scaled = gradient_descent(X_train, y_train, alpha=0.01, num_iterations=1000)

# Make predictions on the test data
y_pred_z_score_scaled = predict(X_test, w_z_score_scaled, b_z_score_scaled)

##########################################################################################################

# Evaluate the model using MSE
mse_max_scaled = calculate_mse(y_test, y_pred_max_scaled)
mse_mean_scaled = calculate_mse(y_test, y_pred_mean_scaled)
mse_z_score_scaled = calculate_mse(y_test, y_pred_z_score_scaled)
print("\n\n")
print(60 * "#" + "\n")
print("Mean Squared Error (Max Scaling):", mse_max_scaled)
print("Mean Squared Error (Mean Scaling):", mse_mean_scaled)
print("Mean Squared Error (Z-Score Scaling):", mse_z_score_scaled)
print("\n" + 60 * "#")
print("\n\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Generate synthetic data with different feature scales
np.random.seed(0)

# Daily temperature (feature x1) with a range of [20, 40] degrees Celsius
temperature = 20 + 20 * np.random.rand(100, 1)

# Ice cream sales (feature x2) with a range of [0, 200] units
ice_cream_sales = 200 * np.random.rand(100, 1)

# Combine temperature and ice cream sales into a single feature matrix X
X = np.hstack((temperature, ice_cream_sales))

# Daily profit (target variable y)
# Assume that daily profit depends on the temperature, ice cream sales,
# and some random noise (normal distribution)
daily_profit = 1000 + 30 * temperature + 5 * ice_cream_sales + np.random.randn(100, 1)

# Max Scaling
def max_scale(X):
    return X / X.max(axis=0)

X_max_scaled = max_scale(X.copy())

# Mean Scaling
def mean_scale(X):
    return (X - X.mean(axis=0)) / X.std(axis=0)

X_mean_scaled = mean_scale(X.copy())

# Z-Score Scaling
def z_score_scale(X):
    return StandardScaler().fit_transform(X)

X_z_score_scaled = z_score_scale(X.copy())


# For Max Scaling
X_train_max, X_test_max, y_train_max, y_test_max = train_test_split(X_max_scaled, daily_profit, test_size=0.2, random_state=42)

# Initialize the weights and bias terms closer to zero
n_features_max = X_train_max.shape[1]
w_max = np.zeros((n_features_max, 1))
b_max = 0


# For Mean Scaling
X_train_mean, X_test_mean, y_train_mean, y_test_mean = train_test_split(X_mean_scaled, daily_profit, test_size=0.2, random_state=42)

# Initialize the weights and bias terms closer to zero
n_features_mean = X_train_mean.shape[1]
w_mean = np.zeros((n_features_mean, 1))
b_mean = 0

# Rest of the code...

# For Z-Score Scaling
X_train_z_score, X_test_z_score, y_train_z_score, y_test_z_score = train_test_split(X_z_score_scaled, daily_profit, test_size=0.2, random_state=42)

# Initialize the weights and bias terms closer to zero
n_features_z_score = X_train_z_score.shape[1]
w_z_score = np.zeros((n_features_z_score, 1))
b_z_score = 0


# Use a smaller learning rate for smoother convergence
alpha = 0.01  # Learning rate
num_iterations = 1000


########################################################################################################
# For Max Scaling
########################################################################################################


# Implement gradient descent
m_max = len(X_train_max)  # Number of training data points

for iteration in range(num_iterations):
    # Compute predictions
    y_pred_max = np.dot(X_train_max, w_max) + b_max

    # Compute gradients
    w_gradient_max = (1 / m_max) * np.dot(X_train_max.T, (y_pred_max - y_train_max))
    b_gradient_max = (1 / m_max) * np.sum(y_pred_max - y_train_max)

    # Update weights and bias
    w_max -= alpha * w_gradient_max.reshape(-1, 1)  # Reshape to match the shape of w
    b_max -= alpha * b_gradient_max

# Make predictions on the test data
y_pred_max = np.dot(X_test_max, w_max) + b_max

# Evaluate the model using Mean Squared Error (MSE)
mse_max = np.mean((y_pred_max - y_test_max) ** 2)
print("Mean Squared Error (Max Scaling):", mse_max)

# Plot one feature (e.g., x1) against y_test_max
plt.scatter(X_test_max[:, 0], y_test_max, label='Test Data Points', alpha=0.7)
plt.xlabel('Feature x1')
plt.ylabel('Daily Profit')
plt.legend()
plt.title('Linear Regression with Gradient Descent - Synthetic Data and Max Scaling')
plt.show()

# Print the learned coefficients
print("Learned Coefficients (Weights) with Max Scaling:")
print(w_max)
print("Bias (Intercept) with Max Scaling:", b_max)
print(50 * "#")
print("\n")

########################################################################################################
# For Mean Scaling
########################################################################################################

# Implement gradient descent
m_mean = len(X_train_mean)  # Number of training data points

for iteration in range(num_iterations):
    # Compute predictions
    y_pred_mean = np.dot(X_train_mean, w_mean) + b_mean

    # Compute gradients
    w_gradient_mean = (1 / m_mean) * np.dot(X_train_mean.T, (y_pred_mean - y_train_mean))
    b_gradient_mean = (1 / m_mean) * np.sum(y_pred_mean - y_train_mean)

    # Update weights and bias
    w_mean -= alpha * w_gradient_mean.reshape(-1, 1)  # Reshape to match the shape of w
    b_mean -= alpha * b_gradient_mean

# Make predictions on the test data
y_pred_mean = np.dot(X_test_mean, w_mean) + b_mean

# Evaluate the model using Mean Squared Error (MSE)
mse_mean = np.mean((y_pred_mean - y_test_mean) ** 2)
print("Mean Squared Error (Mean Scaling):", mse_mean)

# Plot one feature (e.g., x1) against y_test_mean
plt.scatter(X_test_mean[:, 0], y_test_mean, label='Test Data Points', alpha=0.7)
plt.xlabel('Feature x1')
plt.ylabel('Daily Profit')
plt.legend()
plt.title('Linear Regression with Gradient Descent - Synthetic Data and Mean Scaling')
plt.show()

# Print the learned coefficients
print("Learned Coefficients (Weights) with Mean Scaling:")
print(w_mean)
print("Bias (Intercept) with Mean Scaling:", b_mean)
print(50 * "#")
print("\n")

########################################################################################################
# For Z-Score Scaling
########################################################################################################


# Implement gradient descent
m_z_score = len(X_train_z_score)  # Number of training data points

for iteration in range(num_iterations):
    # Compute predictions
    y_pred_z_score = np.dot(X_train_z_score, w_z_score) + b_z_score

    # Compute gradients
    w_gradient_z_score = (1 / m_z_score) * np.dot(X_train_z_score.T, (y_pred_z_score - y_train_z_score))
    b_gradient_z_score = (1 / m_z_score) * np.sum(y_pred_z_score - y_train_z_score)

    # Update weights and bias
    w_z_score -= alpha * w_gradient_z_score.reshape(-1, 1)  # Reshape to match the shape of w
    b_z_score -= alpha * b_gradient_z_score

# Make predictions on the test data
y_pred_z_score = np.dot(X_test_z_score, w_z_score) + b_z_score

# Evaluate the model using Mean Squared Error (MSE)
mse_z_score = np.mean((y_pred_z_score - y_test_z_score) ** 2)
print("Mean Squared Error (Z-Score Scaling):", mse_z_score)

# Plot one feature (e.g., x1) against y_test_z_score
plt.scatter(X_test_z_score[:, 0], y_test_z_score, label='Test Data Points', alpha=0.7)
plt.xlabel('Feature x1')
plt.ylabel('Daily Profit')
plt.legend()
plt.title('Linear Regression with Gradient Descent - Synthetic Data and Z-Score Scaling')
plt.show()

# Print the learned coefficients
print("Learned Coefficients (Weights) with Z-Score Scaling:")
print(w_z_score)
print("Bias (Intercept) with Z-Score Scaling:", b_z_score)
```
